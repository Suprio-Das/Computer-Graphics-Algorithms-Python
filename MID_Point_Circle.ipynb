```python
import matplotlib.pyplot as plt

def mid_point_circle(radius):
    x = 0
    y = radius
    d = 1 - radius
    points = []

    while x <= y:
        points.extend([(x, y), (y, x), (y, -x), (x, -y), (-x, -y), (-y, -x), (-y, x), (-x, y)])
        if d < 0:
            d += 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1
        x += 1

    return points

def draw_circle(radius):
    points = mid_point_circle(radius)
    x_coords, y_coords = zip(*points)

    plt.scatter(x_coords, y_coords, color='blue')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Mid-Point Circle Algorithm")
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Example usage
draw_circle(10)
```
