```python
import matplotlib.pyplot as plt

def mid_point_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    d = 2 * dy - dx
    y = y1
    points = [(x1, y1)]

    for x in range(x1, x2 + 1):
        points.append((x, y))
        if d < 0:
            d += 2 * dy
        else:
            y += 1
            d += 2 * (dy - dx)

    return points

def draw_line(x1, y1, x2, y2):
    points = mid_point_line(x1, y1, x2, y2)
    x_coords, y_coords = zip(*points)

    plt.plot(x_coords, y_coords, marker='D',markersize = 10, color='black', linestyle='dashed')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Mid-Point Line Algorithm")
    plt.grid()
    plt.show()

draw_line(2, 3, 10, 8)
```
